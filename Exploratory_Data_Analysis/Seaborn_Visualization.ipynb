### Preliminaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#seaborn settings
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8,4)})
```

### Load Data Sets

```python
titanic = pd.read_csv('data/titanic.csv')
```

```python
server = pd.read_csv('data/serverdata.csv')
```

```python
mlb = pd.read_csv('data/mlbsalaries.csv')
```

```python
titanic.head()
```

```python
server.head()
```

```python
mlb.head()
```

### 1: Histograms

```python
plt.hist(titanic.Age.dropna(),bins=25)
plt.show()
```

```python
plt.hist(server.Time, 25, color=sns.desaturate("indianred",1))
plt.show()
```

```python
plt.hist(mlb.Salary.dropna(), 25)
plt.show()
```

### 2: Boxplot

```python
sns.boxplot(titanic.Age.dropna())
plt.show()
```

```python
sns.boxplot(titanic.Age.dropna(), titanic.Sex)
plt.show()
```

```python
sns.boxplot(server.Time)
plt.show()
```

```python
sns.boxplot(mlb.Salary)
plt.show()
```

### 3: Kernel Density Estimate Plot

```python
sns.kdeplot(titanic.Age.dropna(), shade=True)
plt.show()
```

```python
sns.kdeplot(server.Time.dropna(), shade=True)
plt.show()
```

```python
sns.kdeplot(mlb.Salary.dropna(), shade=True)
plt.show()
```

```python
sns.distplot(titanic.Age.dropna())
plt.show()
```

```python
sns.distplot(server.Time.dropna())
plt.show()
```

```python
sns.distplot(mlb.Salary.dropna())
plt.show()
```

### 4: Violin Plot

```python
sns.violinplot(titanic.Age.dropna())
plt.show()
```

```python
sns.violinplot(titanic.Age.dropna(), titanic.Sex)
plt.show()
```

```python
sns.violinplot(mlb.Salary)
plt.show()
```

### 5: Cumulative Distribution Function

```python
sns.kdeplot(titanic.Age.dropna(), cumulative=True)
plt.show()
```

```python
sns.kdeplot(server.Time.dropna(), cumulative=True)
plt.show()
```

```python
sns.kdeplot(mlb.Salary.dropna(), cumulative=True)
plt.show()
```
